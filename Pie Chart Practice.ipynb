```python
import matplotlib.pyplot as plt
```

```python
categories = ["pants", "shirts", "outerwear", "pajamas", "underwear", "shoes"]
```

```python
revenue = [25, 15, 5, 20, 10, 25]
```

```python
plt.pie(revenue)
plt.show()
```

```python
plt.pie(revenue, labels = categories)
plt.show()
```

```python
plt.pie(revenue, labels = categories, autopct = "%1.1f%%" )
plt.show()
```

```python
explode = [0, .1, 0, 0, 0, 0]
```

```python
plt.pie(revenue, labels = categories, autopct = "%1.1f%%", explode = explode, shadow = True, startangle = 140)
plt.show()
```

```python
# Data info for the pie chart.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create a Pie Chart Using the MATLAB Method
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
plt.subplots(figsize= (8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, labels = x_axis, explode = explode_values, autopct = "%1.1f%%")
plt.show()
```

```python
# Assign 12 colors, one for each month.
plt.subplots(figsize = (8,8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.pie(y_axis, labels = x_axis, explode = explode_values, autopct = "%.1f%%", colors = colors) 
plt.show()
```

```python
# Create a Pie Chart Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.pie(y_axis, labels = x_axis)
plt.show()
```
