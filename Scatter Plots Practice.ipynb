```python
import matplotlib.pyplot as plt
```

```python
x = [50, 42, 39, 34, 32, 48, 22, 17, 8, 13]
y = [17, 19, 21, 22, 24, 23, 34, 31, 43, 35]
```

```python
plt.scatter(x, y , c = "green", marker = "x", s = x)
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Average Fare vs Total Rides")
plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# If we add an "o" at the end of the plt.plot() it will change from plot to scatter.
plt.plot(x_axis, y_axis, 'o')
```

```python
# This is the direct way or how to do the scatter plot.
plt.scatter(y_axis, x_axis, c = "red", label = "Chicago")
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.title("PyBer Fare By Month")
plt.legend()
plt.xlim(0, 50)
plt.gca().invert_yaxis()
```

```python
# Creating a bubble chart using MATLAB method.
plt.scatter(x_axis, y_axis)
```

```python
# Setting the y_axis smaller by using a for loop.
plt.scatter(x_axis, y_axis)
y_axis_larger = []
for Data in y_axis:
    y_axis_larger.append(Data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# The format for Python's list comprehension is as follows.
new_list = [expression for item in list if conditional]
```

```python
# Setting the y_axis smaller using the Python's list comprehension technique
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create a Scatter Plot Using the Object-Oriented Interface
fig, ax = plt. subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a Bubble Chart Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s = y_axis)
```
